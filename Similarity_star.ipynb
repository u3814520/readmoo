#### 使用者評論數大於2 

```python
import pandas as pd
import os 
import json
df=pd.read_json('./userstar.json',lines=True)
df.head()
```

```python
df=df.dropna(axis=0)  # 刪除空值
```

```python
df=df.drop(df[df["ISBN"]=='0'].index,axis=0)   # 刪除 ISBN等於 0
```

```python
#df.loc[df['USERSTAR']==0,'USERSTAR']=float(1.0)    # 評分為 0 分給予 1 分
```

```python
df['ISBN'] = df['ISBN'].astype('str')  # 確認 ISBN 為 STR
df['USERSTAR'] = df['USERSTAR'].astype('float64')  # 確認USERSTAR 為 float64
```

```python
df_star=df
df_star
```

```python
#評論數大於2 抓出
df_star=df_star.groupby('USER').filter(lambda x : len(x)>1)
df_star=df_star.reset_index(drop=True)
```

```python
df_star['user_id'] = pd.Categorical(df_star.USER).codes 
```

```python
df_star=df_star.sort_values(by=['user_id']) # 為使用者上ID
df_star
```

```python
df_starpivot = df_star.pivot_table(index='user_id', columns='ISBN', values='USERSTAR')
df_starpivot
```

```python
df_newpivot= df_starpivot.fillna(float(0))  # 空值補 0
```

```python
to_array=df_newpivot.values   # 轉成向量
to_array
```

```python
users=df_newpivot.index.astype('int').max()   # 轉型別
userid=[i for i in range(users+1)]
```

```python
from sklearn.metrics.pairwise import cosine_similarity   # 餘絃相似度
cos_sims = cosine_similarity(to_array)
sims_df=pd.DataFrame(cos_sims,columns=userid,index=userid)
sims_df
```

```python
# 取出 user_id 和 similarity 做成 list
sim_scores=list(sims_df.iloc[:,33].nlargest(5))
# 與 user_id 與 其他使用者的相似度
sim_users=list(sims_df.iloc[:,33].nlargest(5).index)
print(sim_users)
sim_scores
```

```python
# 丟入dictionary
sims_dict={}
for i in range(1,5):
    sims_dict[str(sim_users[i])]=sim_scores[i]
print(sims_dict)
```

```python
# 比對差集(雙方沒有交集的書籍) 減 自己看過的書 = 對方和自己沒交集的書
own=list(df_star.loc[df_star.user_id==33]['BOOKNAME'])
print(own)
for k,v in sims_dict.items():
    the_other=list(df_star.loc[df_star.user_id==int(k)]['BOOKNAME'])
    print(f'還可以看{set(the_other)-set(own)}，讀者{k}相似度{v:.2f}')
    print()
```
