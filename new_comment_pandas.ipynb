```python
 import statsmodels.api as sm
df2['user_id'] = pd.Categorical(df2.USER).codes   
df2
```

```python
userBOOK = df2.pivot_table(index=['user_id'],columns=['BOOKNAME'],values='USERSTAR')
userBOOK.head()
```

```python
corrBOOK = userBOOK.corr()
corrBOOK.head()
# starBOOK = userBOOK[9789869899901]
# starBOOK
```

```python
myBOOK = userBOOK.loc[7885].dropna()
myBOOK
```

```python
simCandidates = pd.Series()
for i in range(0, len(myBOOK.index)):
#     print ("Adding sims for " + myBOOK.index[i] + "...")
    # 找出我評分過的書
    sims = corrBOOK[myBOOK.index[i]].dropna()
    # 放大我對書評分的相似度 給予權重
    sims = sims.map(lambda x: x * myBOOK[i])
    # 將分數加到相似度候選串列中
    simCandidates = simCandidates.append(sims)
    
# 看一下目前結果:
print ("sorting...")
simCandidates.sort_values(inplace = True, ascending = False)
print (simCandidates.head(10))
```
