```python
# 讀出 readmoo 有評分的檔案
import pymongo
import pandas as pd
from pymongo import MongoClient
MONGO_HOST = '10.2.14.10'
MONGO_DB = 'userstar'
MONGO_COLLETION = 'userstar'

def connect_mongo():  #連線資料庫
    global collection
    client = MongoClient(MONGO_HOST, 27017)
    db = client[MONGO_DB]
    collection = db[MONGO_COLLETION]

connect_mongo()  #呼叫連線資料庫函式
cursor = collection.find()  #依query查詢資料
userstar_df=  pd.DataFrame(list(cursor))  #轉換成DataFrame
del userstar_df['_id']
userstar_df['ISBN'] = userstar_df['ISBN'].astype('str')  # 確認 ISBN 為字串
userstar_df.head()
```

```python
# 使用 snownlp文本
from snownlp import SnowNLP,sentiment
sentiment.train(r'C:\Users\Tibame\Desktop\readmoo\neg1.txt', r'C:\Users\Tibame\Desktop\readmoo\pos1.txt')
```

```python
# 挑出 ISBN 為 0
readroom_df=userstar_df[userstar_df["USERSTAR"]==0]
readroom_df=readroom_df.dropna()
```

```python
# 挑出 ISBN 不為 0
readroom_df1=userstar_df[userstar_df["USERSTAR"]!=0]
readroom_df1=readroom_df1.dropna()
```

```python
# snownlp 給分規則
def snow_result(CONTENT):
    s = SnowNLP(CONTENT)
    if s.sentiments >= 0.8:
        return float(5.0)
    elif s.sentiments >= 0.6:
        return float(4.0)
    elif s.sentiments >= 0.3:
        return float(3.0)
    else:
        return float(1.0)
```

```python
# readoom給分
readroom_df['USERSTAR'] = readroom_df.CONTENT.apply(snow_result)
readroom_df
```

```python
# king給分
king_df['USERSTAR'] = king_df.CONTENT.apply(snow_result)
king_df
```

```python
# 合併
userstar_df = pd.concat([readroom_df,readroom_df1],axis=0)
userstar_df
```

```python
# 合併
df = pd.concat([userstar_df,king_df],axis=0)
df
```
